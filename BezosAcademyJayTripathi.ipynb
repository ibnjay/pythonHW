**B) What considerations went into your database build (eg, why did you choose the data model, database type and instance settings?). Articulate any assumptions you made in determining those choices.**

#### Data Model

Data model chosen for its simplicity and lack of knowledge about what other data we would want to keep in future.


Provided more time : I would choose to create a  primary key for combination of `state`, `county` and `tract` and create an ID for that combination. Create a table with these 4 fields that identify unique combination of `state`, `county` and `tract`.  Example : 
```CREATE TABLE state_country_tract (
    sct_id serial UNIQUE,
    state TEXT,
    country TEXT,
    tract TEXT,
    PRIMARY KEY(state, country, tract)
)
```

Then all future conesus variable or combination of such variables that becomes a table doesn't need to store `state`, `county` and `tract` in tables. They just store the sct_id  and can reference pull `state`, `county` and `tract` from `state_country_tract` table with a join.


#### database type and instance settings

**Database type**:
I chose a postgres RDS, its free and widely used. Doesn't tie us to a cloud provider. Not as scalable as Arora, but my assumption is current data size is not very large. I didn't consider a noSQL data store because of usiblity and learning curve associated with it for other users.

**Instance setting**:
Currently chose `t3.micro` as it's free. In prod scenario would start with a medium, and scale up when it becomes necessary (assuming inplace vertical scaling is very easy). In prod scenario would turn on backup and snapshot, specify maintainace windows, utilize monitoring. Current RDS does use KMS encryption
